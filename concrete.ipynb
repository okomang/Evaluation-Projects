# SCALING DATA INTO X AND Y TO FIT IN MACHINE LEARNING MODELS

```python
x=C.iloc[ : , :-1]
y=C.iloc[ : ,-1]
```

```python
x
```

```python
y
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
def max_r2(reg,x,y):
    max_r_score=0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=.20)
        reg.fit(x_train,y_train)
        pred=reg.predict(x_test)
        r2_scr=r2_score(pred,y_test)
        print('R2 Score is',r2_scr,'at Random State',r_state)
        
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    
    print('\n')
    print('Max R2 Score is',max_r_score, 'at random state',final_r_state)
```

```python
lr=LinearRegression()
r_state=max_r2(lr,x,y)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn import svm
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=100,test_size=0.20)
```

```python
knr=KNeighborsRegressor()
parameters={ 'n_neighbors' : range(0,15)}
g=GridSearchCV(knr,parameters,cv=10)
g.fit(x_train,y_train)
g.best_params_
```

```python
svr=SVR()
parameters={'kernel' : ['poly','rbf','linear']}
g=GridSearchCV(svr,parameters,cv=10)
g.fit(x_train,y_train)
g.best_params_
```

```python
knr=KNeighborsRegressor(n_neighbors=2)
r_state=max_r2(knr,x,y)
```

```python
svr=SVR(kernel='rbf')
r_state=max_r2(svr,x,y)
```

```python
dt=DecisionTreeRegressor()
r_state=max_r2(dt,x,y)
```

Using Regularization Model Lasso

```python
from sklearn.linear_model import Lasso
l=Lasso()
parameters={ 'alpha' : [0.0001,0.001,0.01,1]}
g=GridSearchCV(l,parameters,cv=10)
g.fit(x_train,y_train)
g.best_params_
```

```python
l=Lasso(alpha=0.0001)
r_state=max_r2(l,x,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
r=RandomForestRegressor()
parameters={'n_estimators' : range(0,21)}
g=GridSearchCV(r,parameters,cv=10)
g.fit(x_train,y_train)
g.best_params_
```

```python
r=RandomForestRegressor(n_estimators=18)
r_state=max_r2(r,x,y)
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada=AdaBoostRegressor()
parameters={'learning_rate':[0.0001,0.001,0.01,1],'base_estimator': [lr,l,knr],'n_estimators':[10,50,100,150]}
g=GridSearchCV(ada,parameters,cv=10)
g.fit(x_train,y_train)
g.best_params_
```

```python
ada=AdaBoostRegressor(base_estimator=knr,learning_rate=0.0001,n_estimators=150)
r_state=max_r2(ada,x,y)
```

```python
model=[]
cvscore=[]
MAE=[]
MSE=[]
from sklearn.model_selection import cross_val_score
MM=[lr,knr,svr,dt,l,r,ada]

for m in MM:
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=100,test_size=0.20)
    m.fit(x_train,y_train)
    pred=m.predict(x_test)
    print('********',m,'************')
    
    mse=mean_squared_error(pred,y_test)
    print('Mean Squared Error is',mse)
    MSE.append(mse)
    
    mae=mean_absolute_error(pred,y_test)
    print('Mean Absolute Error is',mae)
    MAE.append(mae)
    
    c=cross_val_score(m,x,y,scoring='r2',cv=10).mean()
    print('CV score is',c)
    cvscore.append(c)
    print('\n')
```

```python
final=pd.DataFrame({'Model':MM,'MAE' : MAE,'MSE': MSE, 'CV SCORE':cvscore,'R2 Score' : [' 0.8190353233400884 at random state 100',
               '0.8983769568403631 at random state 69','0.9107008576743956 at random state 51','0.9035913056652797 at random state 89',
                  '0.8189493783303426 at random state 100','0.925638592707619 at random state 83','0.9009604978088566 at random state 69']})
```
