```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
def r(regr,x,y):
    max_r_score=0
    final_r_state=0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        pred=regr.predict(x_test)
        r2=r2_score(pred,y_test)
        
        print('R2 Score is',r2,'at random state',r_state)
        
        if r2>max_r_score:
            max_r_score=r2
            final_r_state=r_state
    
    print('\n')
    print('Max R2 score is',max_r_score,'At Random State',final_r_state)
    return(final_r_state)
```

```python
lr=LinearRegression()
r_state=r(lr,x,y)
```

```python
dt=DecisionTreeRegressor()
r_state=r(dt,x,y)
```

```python
model=[]
r2=[]
mse=[]
mae=[]
cv=[]

Model=[lr,dt,knr,l]

for m in Model:
    print(m)
    
    m.fit(x_train,y_train)
    pred=m.predict(x_test)
    
    
    
    r_2=r2_score(pred,y_test)
    r2.append(r_2)
    print('R2 Score is',r_2)
    
    
    Mae=mean_absolute_error(pred,y_test)
    print('Mean Absolute Error is',Mae)
    Mse=mean_squared_error(pred,y_test)
    print('Mean Sqaured Error is',Mse)
    
    c=cross_val_score(m,x,y,cv=10,scoring='r2').mean()
    print('Cross Value Score is',c)
    cv.append(c)
    mae.append(Mae)
    mse.append(Mse)
    print('\n')
```
