This is a restaurant Dataset.
Giving us insights of what kind of food is consumed by people and what is the cost incurred on it.
Having the following columns:
    TITLE: The feature of the restaurant which can help identify what and for whom it is suitable for.

RESTAURANT_ID: A unique ID for each restaurant.

CUISINES: The variety of cuisines that the restaurant offers.

TIME: The open hours of the restaurant.

CITY: The city in which the restaurant is located.

LOCALITY: The locality of the restaurant.

RATING: The average rating of the restaurant by customers.

VOTES: The overall votes received by the restaurant.

COST: The average cost of a two-person meal.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
def r(regr,x,y):
    max_r_score=0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        pred=regr.predict(x_test)
        r2=r2_score(pred,y_test)
        
        print('R2 Score is',r2,'at random state',r_state)
        
        if r2>max_r_score:
            max_r_score=r2
            final_r_state=r_state
    
    print('\n')
    print('Max R2 score is',max_r_score,'At Random State',final_r_state)
    return(final_r_state)
```

```python
lr=LinearRegression()
r_state=r(lr,x,y)
```

```python
dt=DecisionTreeRegressor()
r_state=r(dt,x,y)
```
