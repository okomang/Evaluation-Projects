```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
```

```python
def max_r2(reg,x,y):
    max_r_score=0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.20)
        reg.fit(x_train,y_train)
        pred=reg.predict(x_test)
        r2=r2_score(pred,y_test)
        print('R2 Score is',r2,'At Random State',r_state)
        
        if r2>max_r_score:
            max_r_score=r2
            final_r_state=r_state
    
    print('\n')
    print('Max R score is',max_r_score,'at Random State',final_r_state)
    return final_r_state
        
```

```python
lr=LinearRegression()
r_state=max_r2(lr,x,y)
```

```python
from sklearn.model_selection import GridSearchCV
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=56,test_size=0.20)
```

```python
def grid(reg):
    g=GridSearchCV(reg,parameters,cv=10)
    g.fit(x_train,y_train)
    g.best_params_
    
    return g.best_params_
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn import svm
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
svr=SVR()
parameters={'kernel': ['rbf','poly','linear']}
grid(svr)
```

```python
knr=KNeighborsRegressor()
parameters={'n_neighbors': range(0,21)}
grid(knr)
```

```python
r=RandomForestRegressor()
parameters={'n_estimators':[10,50,100,150],'random_state':[56]}
grid(r)
```

```python
gb=GradientBoostingRegressor()
parameters={'learning_rate':[0.001,0.01,0.1],'n_estimators' : [10,50,100,150]}
grid(gb)
```

```python
svr=SVR(kernel='linear')
r=RandomForestRegressor(n_estimators=150)
gb=GradientBoostingRegressor(learning_rate=0.1,n_estimators=50)
knr=KNeighborsRegressor(n_neighbors=2)
```

```python
models=[]
r2=[]
cv=[]
MAE=[]
MSE=[]

model=[lr,svr,r,gb,knr]

for m in model:
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=56,test_size=0.20)
    m.fit(x_train,y_train)
    pred=m.predict(x_test)
    print('*******',m,'*******')
    models.append(model)
    
    r2scr=r2_score(pred,y_test)
    print('R2 Score is',r2scr)
    r2.append(r2scr)
    
    mae=mean_absolute_error(pred,y_test)
    print('MAE is',mae)
    MAE.append(mae)
    
    mse=mean_squared_error(pred,y_test)
    print('MSE is', mse)
    MSE.append(mse)
    
    c=cross_val_score(m,x,y,cv=10).mean()
    print('Cross Val Score is',c)
    cv.append(c)
    
    print('\n')
    
    
```

```python
Final=pd.DataFrame({'Model': model,'R2Score':r2,'MAE':MAE,'MSE':MSE,'CVSCORE':cv})
```
